```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
cmap = ["blue", "black", "red", "green"]

fig = plt.figure(tight_layout=True)

for i in range(4):
    ax = fig.add_subplot(2,2,i+1)
    
    x = np.arange(1, len(dataset["data"])+1)
    y = np.array([d[0] for d in dataset["data"]])
     
    ax.scatter(x, y, marker="o", s=10, color=cmap[i], label=dataset["feature_names"][i])
    ax.set_xlabel("id")
    ax.set_ylabel(dataset["feature_names"][i])
    ax.legend(fontsize=8)
    
fig.suptitle("iris dataset features", y=0)
```
